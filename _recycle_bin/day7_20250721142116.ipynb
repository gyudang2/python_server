```python
# 요청을 위한 라이브러리
import requests
url_news = "https://news.nate.com/recent?mid=n0105"
response = requests.get(url_news)

# request_status_code
# 200 : 요청 성공
# 403 : Forbidden 접근금지                   # 4xx -> client 문제, 5xx -> server 문제
# 404 : Not found 페이지 없음
# 500 : Internal Server Error 서버 오류
```

```python
# 응답 처리하는 3가지 방법
# 1. 문자열
response.text
# 2. byte (이미지, 파일)
response.content
# 3. json (json타입)
response.json() # json 타입만 열기
```

```python
html_news = response.text

# 정적인 웹 페이지으 태그 등에 접근하여
# 내용물을 출력, 처리할 수 있는 
# BeautifulSoup 라이브러리 사용하기

from bs4 import BeautifulSoup as bs

soup_news = bs(html_news, "html.parser")
# soup.find(태그명)
# 처음 보이는 h2 태그를 가져옴
soup_news.find("h2")

# soup.find_all(태그명)
# 모든 태그를 가져옴
soup_news.find_all("h2")[1]


# 가져온 태그.text ->  태그 안의 내용물(content)만 가져옴
soup_news.find_all("h2")[1].text

# for문을 활용하여 모든 최신 IT 뉴스 타이틀 출력
news_list = soup_news.find_all("h2")

for news in news_list:
    # 기사 제목 중에 "AI"가 들어간 기사 제목만 출력하기
    # AI 혹은 SKT가 들어가면 출력
    if "AI" in news.text or "SKT" in news.text:
        print(news.text)


```

```python
# 요청 방식 
# 1. GET 요청 (쿼리스트링)
# key + value를 url에 담아 요청하는 방식
# 뉴스url?키값1=밸류1&키값2=밸류2&키값3=밸류3 ~~~~~

# 장점 : url을 통해 페이지 공유 가능 (같은 정보를 받아올 수 있다.)
# 단점 : 상대적으로 보안이 취약하다, 데이터 전송 용향 제한 있음(url 길이만)

# 2. POST 요청
# key + value를 body에 담아 요청하는 방식
# 장점 : 상대적으로 보안이 강하다, 데이터 전송 용량 제한 없음(파일, 이미지도 전송 가능)
# 단점 : 페이지 공유 불가
```
