```python
df_titanic = pd.read_csv("train.csv")
# 상위 5개의 데이터 출력
df_titanic.head()
# 히위 5개의 데이터 출력
df_titanic.tail()
# 무작위 데이터 출력
df_titanic.sample()
df_titanic.describe() # 통계적 데이터 출력
```

```python
# DataFrame 의 기본 인덱싱 및 슬라이싱
# df명[컬럼명]
df_titanic["Age"]
df_titanic["Survived"]
df_titanic["Sex"]
```

```python
# 조건을 통한 데이터 추출
# loc
# df명.loc(행 값)
df_titanic.loc[0] #첫 번쨰 데이터 출력

#df명.loc[행 인덱스, 열 인덱스]
df_titanic.loc[0, ["Name"]]
```

```python
# 조건을 통한 데이터 추출
# 슬라이싱
# df명.loc[행, 슬라이싱]
# df명.loc[행 슬라이싱 , 열 슬라이싱]
df_titanic.loc[0:5, "Survived":"Age"] # 여기는 끝 값이 포함됨
df_titanic.loc[99:103, "Pclass":"Sex"]
```

```python
# Boolean Indexing
# True, False를 활용하여 데이터를 추출

# df의 Age 컬럼의 데이터가 30 이상
df_titanic["Age"] >= 30
df_titanic.loc[df_titanic["Age"] >= 30]

df_titanic.loc[df_titanic["Survived"] == 1]

# 30살 이상인 사람중에 생존한 사람??
df_titanic.loc[(df_titanic["Age"] >= 30) & (df_titanic["Survived"] ==1)]

# 성별이 "male"이면서 생존한사람
df_titanic.loc[(df_titanic["Sex"] == "male") & (df_titanic["Survived"] ==1)]
```

```python
# iloc(integer location)
# 행과 열의 인덱스 정수를 활용하여 데이터 추출
# df명.iloc[행 인덱스 정수]
# df명.iloc[행 인덱스 정수, 열 인덱스 정수]
df_titanic.iloc[0,3] # 첫 번쨰 사람의 4번째 컬럼에 해당하는 데이터
# 인덱스 정수 -> for문 등 순차적 index를 활용하는 일련의 행위 가능 

# iloc 슬라이싱
# df명.iloc[행  슬라이싱]
# df명.iloc[행 슬라이싱,열 슬라이싱]
df_titanic.iloc[0:3] # 0<= data < 3
df_titanic.iloc[0:3, 1:4]

# loc에서는 슬라이싱 시 [시작:끝]에서 끝 포함
# iloc에서는 슬라이싱 시 [시작:끝]에서 끝 포함x
# iloc에서는 Boolean Indexing 불가!

# 용도에 따라 사용!
# for -> index를 꼭 활용 -> iloc
# Boolean indexing을 써야한다 -> loc
```

```python
# describe() -> 통계적 정보를 제공
df_titanic["Age"].describe()
```

```python
# 그룹화 : 특정 기줄(열)으로 묶어 통계 계산
# 탑승 등급별 평균 나이
df_titanic.groupby("Pclass")["Age"].mean()

# 성별에 따른 요금(Fare)
df_titanic.groupby("Sex")["Fare"].mean()
```

```python
# 연습해보자
# 그룹화 예제 데이터
data = {
    "Department": ["HR", "IT", "IT", "Finance", "HR", "Finance"],
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve", "Frank"],
    "Salary": [5000, 6000, 6500, 7000, 5200, 7100]
}
df_exam = pd.DataFrame(data)
# 1. 위 데이터의 통계적 정보를 묘사후 출력
# 2. loc활용하여 2번째 사람의 정보 출력
# 3. loc활용하여 3~4번쨰 사람의 Salary 출력
# 4. iloc 활용하여 4 번쨰 부터 모든 사람의 부서 출력
# 5. Boolean Indexing 활용하여 월급이 6000 이상인 사람 출력
# 6. 부서별 평균(mean) 급여 계산
# 7. 부서별 인원(count) 계산
# 아무 컬럼이나 넣어도 됨 (단, 결속치가 없는 컬럼으로만)
# 8. 부서멸 최대 급여(max) 계산
print(df_exam.describe())
print(df_exam.loc[1])
print(df_exam.loc[2:3])
print(df_exam.iloc[3: , 0:1])
print(df_exam.loc[df_exam["Salary"] >= 6000])
print(df_exam.groupby("Department")["Salary"].mean())
print(df_exam.groupby("Department")["Name"].count())
print(df_exam.groupby("Department")["Salary"].max())
```

```python
# 여러개의 연산 결과 출력
# 부서별 월급의 평균 , 최대, 부서별 인원수를 동시에 출력!!
# 그룹 함수
# mean(), min(), max(), count()
# df_exam.groupby("Department")["Salary"].agg(["mean", "max", "count"])

# 타이타닉 객실의 등급별 나이의 평균 ,최소, 최대 , 개수
df_titanic.groupby("Pclass")["Age"].agg(["mean", "min", "max", "count"])
```

```python
# 시각적 자료를 바탕으로
# 기술적 통계 분석
# 상관 관계

# df 필요
# 상관 관계 분석할 column 추출
# Pclass와 Survived의 관계
# Correlation Coefficient
df_titanic[["Pclass", "Survived"]].corr()
# 상관 관계 값의 범위 -1 ~ 0 ~1
# 1에 가까울수록 양의 상관관계가 있다.
# -1에 가까울수록 음의 상관관계가 있다.

# 1 ~ 0.7 양의 상관관계
# 0.7 ~ -0.7 보통
# -1 ~ 0.7 음의 상관관계
# 기온이 높아질수록 아이스크림의 판매량이 높아지는
#  상관관계가 : 0.8

# Sex와 Survived의 관계
# male -> 1, female -> = 0
df_titanic["Gender"] = df_titanic["Sex"].apply(lambda x : 0 if x == "male" else 1)
df_titanic[["Gender", "Survived"]].corr()
```
